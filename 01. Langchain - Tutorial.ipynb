```python
from langchain import PromptTemplate
from langchain.chat_models import ChatOpenAI

chatgpt = ChatOpenAI(model_name = 'gpt-3.5-turbo-0613',openai_api_key='Your API Key')
```

```python
chatgpt = ChatOpenAI(model_name = 'gpt-3.5-turbo-0613',openai_api_key='Your API Key', temperature = 0)
answer = chatgpt.predict('가장 유명한 한국음식은 무엇인가요? 50글자 이내로 설명해주세요.')
print(answer)
```

```python
chatgpt = ChatOpenAI(model_name = 'gpt-3.5-turbo-0613',openai_api_key='Your API Key', temperature = 1)
answer = chatgpt.predict('가장 유명한 한국음식은 무엇인가요? 50글자 이내로 설명해주세요.')
print(answer)
```

```python
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler

chatgpt = ChatOpenAI(model_name = 'gpt-3.5-turbo-0613',  streaming = True, openai_api_key='Your API Key', callbacks = [StreamingStdOutCallbackHandler()], temperature = 1)
answer = chatgpt.predict('Why python is the most popular language? answer in korean')
print(answer)
```

```python
from langchain.schema import AIMessage, SystemMessage, HumanMessage

message = [
    SystemMessage(
        content = 'You are a helpful assistant that translate English to Korean.'
    ),
    HumanMessage(
        content = 'I love langchain.'
    )
]

response_langchain = chatgpt(message)
response_langchain
```

```python
chat = ChatOpenAI(model_name = 'gpt-3.5-turbo-0613',  streaming = True, openai_api_key='Your API Key', callbacks = [StreamingStdOutCallbackHandler()], temperature = 1)

answer = chat(
    [
        SystemMessage(content = "당신은 공부 계획을 세워주는 AI 스터디 플래너 입니다. 사용자의 공부 주제를 입력받으면, 이를 학습하기 위한 공부 계획을 작성합니다."),
        HumanMessage(content = "한달동안  Large Language Model에 대해서 공부하고 싶어요.")
    ]
)

print(answer.content)
```
