```python
from langchain.prompts.prompt import PromptTemplate
```

```python
template = """
너는 요리사야. 내가 가진 재료들을 갖고 만들 수 있는 요리 추천하고, 그 요리의 레시피를 제시해줘.
내가 가진 재료는 아래와 같아.

<재료>
{재료}

"""

prompt_template = PromptTemplate(
    input_variables = ['재료'],
    template = template
)

print(prompt_template.format(재료 = '양파, 계란, 사과, 빵'))
```

- Human_message_prompt가 사용자의 입력공간으로, ChatPromptTemplate라는 객체로 값을 전달한다. 이 때, SystemMessagePrompt, HumanMessagePrompt 2가지가 필요하다.

```python
from langchain.prompts import ChatPromptTemplate, PromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate, AIMessagePromptTemplate
from langchain.schema import AIMessage, HumanMessage, SystemMessage
```

```python
system_message_prompt = SystemMessagePromptTemplate.from_template(template)

human_template = '{재료}'
human_message_prompt = HumanMessagePromptTemplate.from_template(human_template)

chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])

answer = chat_gpt(chat_prompt.format_prompt(재료='양파, 계란, 사과, 빵').to_messages())
print(answer.content)
```

## Few show Learning

example(List dict) 로 예제를 전달한다. example_prompt는 PromptTemplate로 input_variables 는 입력값과 출력값, template는 전체 형태를 나타낸다. 입력값과 출력값을 전달하며, FewShowPrompt의 examples로 지정하며, suffix는 입력 형태, input_variables는 입력값을 넣는다.

```python
from langchain.prompts.few_shot import FewShotPromptTemplate

examples = [
    {
        "question": "아이유로 삼행시 만들어줘",
        "answer":
        """
        아: 아이유는
        이: 이런 강의를 들을 이
        유: 유가 없다.
        """
    },
    {
        "question": "김민수로 삼행시 만들어줘",
        "answer":
        """
        김: 김치는 맛있다
        민: 민달팽이도 좋아하는 김치!
        수: 수억을 줘도 김치는 내꺼!
        """
    },
]

example_prompt = PromptTemplate(input_variables = ['question', 'answer'], template = "Question: {question}\n{answer}")

print(example_prompt.format(**examples[0]))
```

```python
prompt = FewShotPromptTemplate(
    examples = examples,
    example_prompt = example_prompt,
    suffix = "Question: {input}",
    input_variables = ["input"]
)

print(prompt.format(input="호날두로 삼행시 만들어줘"))
```

```python
chat_gpt.invoke(prompt.format(input='호날두로 삼행시 만들어줘'))
```

- Fewshot 예제를 동적으로 사용하고 싶을 때, Example Select 사용하며, LLM 여러 작업에서 수행하도록 만들 되, 원하는 범위의 대답을 출력하도록 해애한다. 예제를 적절한 예시만 포함하도록 입력하고, 프롬프트의 입력 길이를 제한하고, 오류가 발생하지 않도록 잘 조절해야 한다.

```python
from langchain.prompts.example_selector import SemanticSimilarityExampleSelector
from langchain.vectorstores import Chroma
from langchain.embeddings import OpenAIEmbeddings
from langchain.prompts import FewShotPromptWithTemplates, PromptTemplate

example_prompt = PromptTemplate(
    input_variables = ['input', 'output'],
    template = "input: {input}\noutput: {output}"
)

examples = [
    {"input": '행복', 'output': "슬픔"},
    {"input": '흥미', 'output': "지루"},
    {"input": '불안', 'output': "안정"},
    {"input": '긴 기차', 'output': "짧은 기차"},
    {"input": '큰 공', 'output': "작은 공"},
]
```

```python
!pip install chromadb tiktoken
```

```python
example_selector = SemanticSimilarityExampleSelector.from_examples(
    # 예시 아이템
    examples,
    # 임베딩 클래스
    OpenAIEmbeddings(),
    # 벡터스토어 클래스, 유사도 검색 및 임베딩 저장
    Chroma,
    # 출력할 예제 갯수
    k=1
)

similar_prompt = FewShotPromptTemplate(
    example_selector = example_selector,
    example_prompt = example_prompt,
    prefix = '주어진 입력에 대해 반대의 의미를 가진 단어를 출력해줘',
    suffix = 'input: {단어}\noutput: ',
    input_variables = ["단어"]
)
```

```python
print(similar_prompt.format(단어='무서운'))
```

```python
print(similar_prompt.format(단어='큰 비행기'))
```

```python
query = '큰 비행기'

print(chat_gpt.invoke(
    similar_prompt.format(단어=query)
))
```

## Output Parser
- 원하는 형태로 결과 출력함.
- CommaSeparatedListOutputParser는 , 로 구분된 리스트 형태로 출력함.
- partital_variables를 통해서 format_instuctions의 값을 주어, 출력 형태로 만듬

```python
from langchain import output_parsers
from langchain.output_parsers import CommaSeparatedListOutputParser
from langchain.prompts import PromptTemplate, ChatPromptTemplate, HumanMessagePromptTemplate
from langchain.chat_models import ChatOpenAI
```

```python
output_parser = CommaSeparatedListOutputParser()

format_instructions = output_parser.get_format_instructions()
format_instructions
```

```python
prompt = PromptTemplate(
    template = "{주제} 5개를 추천해줘.\n{format_instructions}",
    input_variables = ['주제'],
    partial_variables = {'format_instructions': format_instructions}
)
```

```python
prompt.format(주제='영화')
```

```python
_input = prompt.format(주제='영화')
output = chat_gpt.invoke(_input)
```

```python
output_parser.parse(output.content)
```

```python
output_parser.parse(output.content.replace("\"", ""))
```
